### [ python & mysql 연동 ]


```python
## 모듈 로딩
import pymysql

print(f'pymysql version: {pymysql.__version__}')
```

```python
## RDBMS 정보
SEVER_IP = 'localhost'
USER_ID  = 'root'
USER_PW  = 'iopl321'
DB_NAME  = 'mydb'
CHARSET  = 'utf8mb4' 
```

```python
## ==============================================================
## DB 연결
## ==============================================================
conn = pymysql.connect( host        = SEVER_IP,
                        user        = USER_ID,
                        password    = USER_PW,
                        database    = DB_NAME,
                        cursorclass = pymysql.cursors.DictCursor)
```

```python
## ==============================================================
## DB - Select 처리
## ==============================================================
ALL_DATA_SQL = "SELECT * FROM user;"

with conn.cursor() as cursor:
    # Read a single record
    sql = "SELECT id, name FROM user WHERE name = %s"
    cursor.execute(sql, ('마징가',))
    result = cursor.fetchall()
    print(result)
```

```python
## ==============================================================
## DB - insert 처리
## ==============================================================
INSERT_DATA_SQL = "insert into user values (11, '베트맨', 23, '부산', '부산시', '여');"

with conn.cursor() as cursor:
    cursor.execute(INSERT_DATA_SQL)
    conn.commit()
```

```python
## ==============================================================
## DB - delete 처리
## ==============================================================
DELETE_DATA_SQL = "delete from user where id = %s"

with conn.cursor() as cursor:
    cursor.execute(DELETE_DATA_SQL, (11,))
    conn.commit()
```

```python
## ==============================================================
## DB - update 처리
## ==============================================================
UPDATE_DATA_SQL = "update user set name = %s where id = %s;"

with conn.cursor() as cursor:
    cursor.execute(UPDATE_DATA_SQL, ('원더우먼', 10))
    conn.commit()
```

```python
## ==============================================================
## DB 연결 해제
## ==============================================================
conn.close()
```
