reindex(), fill_vaule(), fillna() <hr>

```python
import pandas as pd

dataSR = pd.Series(['A', 'B', 'C'], index=['하나', '둘', '셋'])
print(dataSR)
# 하나    A
# 둘     B
# 셋     C

# 인덱스를 재지정
dataSR2 = dataSR.reindex(['셋', '하나', '넷'], fill_value=1)
dataSR2.fillna(0, inplace= True)
print(dataSR2)
```

```python
## 중복 행 처리
## 데이터 준비
df = pd.DataFrame({ 'brand' : [ 'Yum', 'Yum', pd.NA, 'Indo', 'Indo' ],
                    'style' : [ 'cup', 'cup', 'cup', 'pack', 'pack' ],
                    'rating' : [4, 4, 3.5, 15, 5] })
display(df)
## 중복체크
result = df.duplicated().sum()


print(f'중복 데이터/행 : {result}개')

## 중복체크 - 마지막 유지, 그 외 중복
result = df.duplicated(keep='last')
result

## 중복체크 - 모두 중복
result = df.duplicated(keep=False)
result
## brand 컬럼만 중복인 행
result = df.duplicated(subset='brand')
result
```

```python
## 결측치 제거

#결측치 존재 인덱스 제거
df.dropna(subset = ['brand'])

#결측치 치환
df.fillna(value=0)

## 앞값으로 치환
display(df.ffill())

## 뒷값으로 치환
display(df.bfill())
```

```python
## groupby

df.groupby('brand')['rating'].median()
```

```python
## 필터
df.groupby('brand').filter(lambda x : x['rating'].sum() >= 9.0)
```
