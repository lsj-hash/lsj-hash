### [ 속성 변경 및 설정 ]
- Series/DataFream에는 많은 속성 존재
- 일부는 읽기/변경 가능한 속성
- 일부는 읽기만 가능한 속성 ==> setter 메서드 X

[속성 읽기/변경 가능한 속성]
***

```python
## 모듈 로딩
import pandas as pd
```

```python
## 임의의 데이터
data_dict = {'a':10, 'b':9, 'c':22}

## 시리즈 인스턴스 생성
dataSR = pd.Series(data_dict)

## 출력
print(dataSR)
print(dataSR.index)
```

```python
## index 속성 변경
## Index(['a', 'b,' 'c']. dtype='object')객체
## 변경 '하나', '둘', '셋'

dataSR.index = ['하나', '둘', '셋']

## => 확인
dataSR , dataSR.index
```

```python
## --------------------------------
## name 속성 변경
## --------------------------------
## 현재 name 속성 확인
print("현재",dataSR.name)

## name 속성 설정
dataSR.name = '점수'

print("변경", dataSR.name)
```

[읽기만 가능/변경 불가한 속성들]
***

```python
## --------------------------------
## dtype 속성 변경
## --------------------------------
## 현재 dtype 속성 확인 => 비공개 속성에서 getter 메서드 존재
print("현재",dataSR.dtype)

## name 속성 설정 => 비공개 속성에서 setter 메서드 미존재
#dataSR.name = 'int16'

print("변경", dataSR.dtype)
display(dataSR)
```

```python
## --------------------------------
## values 속성 변경
## --------------------------------
## 현재 dtype 속성 확인 => 비공개 속성에서 getter 메서드 존재
print("현재",dataSR.index)

## values 속성 설정 => 비공개 속성에서 setter 메서드 미존재
# dataSR.values = [0, 9, 11]
display(dataSR)
```

[DataFrame의 속성 읽기/변경]

```python
## 데이터
data_dict = {'번호' : [1,5,9],
             '이름' : ['홍', '마', '베']}

## 데이터 프레임 생성 : key => 컬럼명, Value => 세로 즉, 컬럼 데이터
dataDF = pd.DataFrame(data_dict)

##
display(dataDF, dataDF.columns, dataDF.index)
```

```python
## ------------------------------------------
## - 인덱스 변경
## ------------------------------------------
print("현재", dataDF.index)

# 변경 ['0, 1, 2'] ==> ['사람1', '사람2', '사람3']
dataDF.index = ['사람1', '사람2', '사람3']

## 확인
print(dataDF)
```

```python
## ------------------------------------------
## - columns 변경
## ------------------------------------------
print("현재", dataDF.columns)

# 변경 ['번호', '이름'] ==> ['ID', 'NAME']
dataDF.columns = ['ID', 'NAME']

## 확인
print(dataDF)
```
