### [파이썬에서 변수와 참조]
- 데이터 저장 문법 : 변수명 = 데이터
- 데이터 저장 의미
    * 힙 메모리 영역에 저장된 데이터(객체)의 주소를 저장 
    * 스택 메모리 영역에 변수명 존재함 => 참조형 변수(reference variable)

```python
a = 10 ## int형 인스턴스 10의 주소를 변수 a에 저장
b = 10 ## int형 인스턴스 10의 주소를 변수 b에 저장
c = a  ## int형 인스턴스 10의 주소를 변수 c에 저장
print('id(a)',id(a))
print('id(b)',id(b))
print('id(c)',id(c))
```

```python
a = 5 
b = a
print('id(a)',id(a))
print('id(b)',id(b))

a += 5              ## a = a+5 => 5 + 5 = 10 : 10의 주소
print('id(a)',id(a))
print('id(b)',id(b))
```

```python
datas = [11,22,33] ## 힙에 저장된 [11, 22, 33]의 주소를 저장

a = datas          ## datas 변수가 가진 값 즉, [11, 22, 33]의 주소를 저장

print('id(datas)', id(datas))
print('id(a )   ', id(a ))
print('----------------------')
print('id(datas[0])  ', id(datas[0]))
print('id(datas[1])  ', id(datas[1]))
print('id(a[0])      ', id(a[0]))
print('id(a[1])      ', id(a[1]))

datas[0] = 44
print('----------원소 변경------------')
print('id(datas)', id(datas))
print('id(a )   ', id(a ))
# print(datas, a, sep='\n')
print('----------------------')
print('id(datas[0])  ', id(datas[0]))
print('id(datas[1])  ', id(datas[1]))
print('id(a[0])      ', id(a[0]))
print('id(a[1])      ', id(a[1]))
```

```python
import pandas as pd

sr00 = pd.Series([11, 22], index=['a','b'])


sr11 = sr00


print('----------변경 전------------')
print('id(sr11) ', id(sr11), sr11.index)
print('id(sr00) ', id(sr00), sr00.index)

sr00.index = [11, 22]
print('----------변경 후------------')
print('id(sr11) ', id(sr11), sr11.index)
print('id(sr00) ', id(sr00), sr00.index)

## = 복사하기 copy()
sr22 = sr00.copy()
print('----------변경 후------------')
print('id(sr22) ', id(sr22), sr22.index)
print('id(sr00) ', id(sr00), sr00.index)

sr00.index = ['하나', '둘']
print('----------변경 후------------')
print('id(sr22) ', id(sr22), sr22.index)
print('id(sr00) ', id(sr00), sr00.index)
```
