### [속성 .iloc[] / .loc[] ]
- .iloc[] : 위치기반 인덱스만 사용 가능
- .loc[] : 

[설정 인덱스가 숫자인 경우]
***

```python
import pandas as pd

sr = pd.Series([99, 78], index=[11,22])
display(sr)

print("1번 원소 선택")
print("99 =>", sr[11], sr.loc[11], sr.iloc[0])
print("78 =>", sr[22], sr.loc[22], sr.iloc[1])
```

[설정 인덱스가 글자인 경우]
***

```python
sr = pd.Series([99, 78], index=['a' ,'b'])
display(sr)

print("1번 원소 선택")
print("99 =>", sr['a'], sr.loc['a'], sr.iloc[0])
print("78 =>", sr['b'], sr.loc['b'], sr.iloc[1])
```
