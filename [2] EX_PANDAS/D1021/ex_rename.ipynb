### [ 행 인덱스/열 이름 부분 변경 ]
- Series/DataFream에는 행 인덱스, 열 이름 부분 변경 지원
- 속성명 변경 => 전체 변경. 일부 변경 X
- 메서드 : rename(index, columns, ..., inplace)
    * 행 인덱스 변경 시 => index 매개변수 사용
    * 열 이름 변경 시 => columns 매개변수 사용
    * 원본 변경 여부 => inplace 매개변수
    * 반환값 => None 또는 DataFrame

[Series의 행인덱스 부분 변경]
***

```python
## 모듈 로딩
import pandas as pd
```

```python
## 임의의 데이터
data_dict = {'a':10, 'b':9, 'c':22}

## 시리즈 인스턴스 생성
dataSR = pd.Series(data_dict)

## 출력
print(dataSR)
print(dataSR.index)
```

```python
## --------------------------------- 
## index 속성 변경 : 전체 변경
## ---------------------------------
## Index(['a', 'b,' 'c']. dtype='object')객체
## 변경 '하나', '둘', '셋'

dataSR.index = ['하나', '둘', '셋']

## => 확인
dataSR 
```

```python
## --------------------------------
## index 속성 변경 : 일부/부분 변경
## --------------------------------
## ['하나', '둘', '셋'] =>변경 '[One', '둘', 'Three']
#dataSR.index = ['One', 'Three'] ## 개수 불일치 X 

##=> rename()메서드
dataSR1 =dataSR.rename(index={'하나': 'ONE','셋' : "Three"})
display(dataSR, dataSR1)

##=> rename(index, inplace=Ture)메서드
dataSR11 = dataSR1.rename(index={'둘': '222'}, inplace=True)
display(dataSR, dataSR11)
```
