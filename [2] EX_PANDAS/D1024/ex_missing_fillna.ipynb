### 【데이터 전처리 - 결측치 치환】
- 결측치 (Missing Value)    
    * 데이터/값이 존재하지 않는 것
- 결측치 체크
    * isna() : 결측치 여부에 따라 True/False 반환
- 결측치 처리
    * 삭제 : dropna()
    * 치환 : fillna()

    
[1] 모듈 로딩 및 데이터 준비 <hr>

```python
## 모듈 로딩
import pandas as pd
import numpy as np
```

```python
## DF 생성
df = pd.DataFrame( { 'age' : [ 5, 6, np.nan ],
                     'born': [ pd.NaT,pd.Timestamp('1939-05-27'), pd.Timestamp('1940-04-25')],
                     'name' : ['Alfred', 'Batman', ''],
                     'toy' : [None, 'Batmobile', 'Joker']})

## DF 출력
display( df )

## 기본 요약 정보
df.info( )

```

```python
## 결측치 체크
display(df.isna())

print(f'컬럼별 결측치 수 \n{df.isna().sum()}')
```

[3] 결측치 치환 : fillna() <hr>

```python
df.fillna(value=0)
```

```python
## ----------------------------------------------------
## 최환값 : 컬럼별로 다른 결측치 => 평균, 중앙값, 최빈값 등등
## -----------------------------------------------------
age_mean = df['age'].mean()
born_mode = df['born'].mode()[0]
toy_mode = df['toy'].mode()[0]

df.fillna({'age':age_mean, 'born': born_mode, 'toy': toy_mode})
```

```python
## ----------------------------------------------------
## 최환값 : 이전값/이후값
## -----------------------------------------------------
## 이전행 값
# df.fillna(method='ffill')
display("다음 행 값", df.ffill)

## 다음행 값
# df.fillna(method='bfill')
display("다음 행 값", df.bfill)
```
