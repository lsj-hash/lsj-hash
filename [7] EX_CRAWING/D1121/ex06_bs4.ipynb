### 【 웹 스크래핑 - 정보 추출 】
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함


[1] 모듈로딩 <hr>

```python
from urllib.request import urlopen      ## WEB에서 URL기반으로 다운로드
from bs4 import BeautifulSoup           ## WEB데이터 객체화 및 정보 추출용 모듈
```

[2] Markup 문자열에서 정보 추출<hr>

- CSS 선택자를 활용해서 Tag 탐색 및 추출
    * select_one() 메서드
        * 조건에 맞는 첫 번째 Tag 반환
        * find() 메서드와 검색 결과 동일
    * select() 메서드
        * 조건에 맞는 모든 Tag List로 반환
        * find_all()메서드와 검색 결과 동일
- CSS 선택자
    * Tag의 id 속성 : #id명
    * Tag의 class 속성 : .class명
    * 자식태그 : 부모태그 > 자식태그
    * 자손태그: 상위/부모태그 하위/자손태그

```python
## HTML파일에서 일부분만 선택 str로 저장
data_str ="""
<html>
<body>
<div id ="meigen">
    <h1>위키북스 도서</h1>
    <ul class ="items">
        <li> 유니티 게임 이펙트 입문</li>
        <li>스위프트로 시작하는 아이폰 앱 개발 교과서</li>
        <li>모던 웹사이트 디자인의 정석</li>
    </ul>
</div>

</body>
</html>
"""
```

```python
## markup str ==> 객체화
mbs = BeautifulSoup(data_str, 'html.parser')
```

```python
## ---------------------------------------------------------------
## [2] select_one() 메서드 활용 : BS객체변수명.select_noe(css_selector_str)
##                      필요한 부분을 CSS 쿼리로 추출
## ---------------------------------------------------------------
## => 1개만 추출 : 타이틀 부분
h1 = mbs.select_one("div#meigen > h1").string
print("h1 -",h1)

## => 모두 추출 : 목록 부분
li_list = mbs.select("div#meigen > ul.items > li")
for li in li_list :
    print("li=",li.string)
```

[실습]  fruits-vegetavles.html 파일에서 CSS Selector로 추출 <hr>

```python
## html 파일 => BS 객체 변환 저장
HTML_FILE = '../data/chap02.html'

with open(HTML_FILE, encoding='utf-8') as f:
    BS = BeautifulSoup(f.read(), 'html.parser')

```

```python
#[1] id가 footer인 h1 태그 추출
BS.select_one('h1#footer')
```

```python
#[2] class가 footer인 h1 태그 추출
BS.select_one('a.internal_link')
```

```python
#[3] 부모Tag>자식Tag로 a 태그 추출
BS.select_one('#link>.external_link')
BS.select_one('#link>.external_link2') # 둘다 상관 X
```

```python
## [4] 상위/조상 Tag        하위/자손Tag 추출
BS.select_one('div#class1  #second')
```

```python
## [5] 부모/자식, 조상/자손 관계 X,  여러개 추출
BS.select('#')
```
