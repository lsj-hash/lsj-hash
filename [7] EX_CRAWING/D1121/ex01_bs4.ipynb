### 【 웹 스크래핑 - 정보 추출 】
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함


[1] 모듈로딩 <hr>

```python
from urllib.request import urlopen      ## WEB에서 URL기반으로 다운로드
from bs4 import BeautifulSoup           ## WEB데이터 객체화 및 정보 추출용 모듈
```

[2] WEB에서 데이터 다운로드<hr>

```python
## 다운로드 받을 웹 사이트 URL 
W_URL = 'http://quotes.toscrape.com/'

## 다운로딩
retObj = urlopen(W_URL)
print(f'retObj : {type(retObj)}')

## 로우데이터만 가져오기 => byte 타입의 데이터
bData=retObj.read()
```

[3] WEB 데이터에서 원하는 데이터 추출 : 스크래핑<hr>

```python
## BS객체 생성 : 마크업언어 스타일 데이터 => 사람이 이해하기 쉽게 객체화
bs = BeautifulSoup(retObj.read(), 'html.parser')
```

```python
## BS객체의 속성들
print(bs.h1)
print('-'*30)
print(bs.div)
print('-'*30)
print(bs.a)
```

[4] html 파일에서 정보 추출<hr>

```python
## HTML파일에서 일부분만 선택 str로 저장
data_str ="""
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>BeautifulSoup 활용</title>
</head>
"""
```

```python
## markup str ==> 객체화
mbs = BeautifulSoup(data_str, 'html.parser')
```

```python
## 브라우저에서 웹페이지 타이틀 데이터 추출
print("title 태그 전체 추출", mbs.title, type(mbs.title))
print("title 태그에서 속성 추출         :", mbs.title.text)
print("title 태그에서 getter메서드 추출 :", mbs.title.get_text())
```
