### 【 웹 스크래핑 - 정보 추출 】
- 수집된 웹 데이터에서 필요한 정보 추출하는 작업 : 스크래핑
- 스크래핑을 편리하게 진행하기 위한 패키지 : BeautifulSoup4
- binary/str 타입의 웹 데이터 ==> 객체화 (DOM)해서 프로그램 쉽게 사용할 수 있도록 함

[1] 모듈 로딩 <hr>

```python
from urllib.request import urlopen          ## WEB에서 URL기반으로 다운로드용 모듈
from bs4 import BeautifulSoup               ## WEB데이터 객체화 및 정보 추출용 모듈
```

[2] Markup 문자열에서 정보 추출<hr>

- Tag사이의 텍스트 읽기
    * Tag객체명.text : <시작태그> 텍스트만 존재 </끝태그> 
    * Tag객체명.string : <시작태그> 텍스트만 존재 <다른태그> ... </다른태그> </끝태그> 

- Tag의 속성 읽기
    * Tag객체명['속성명'] 
    * Tag객체명.get('속성명')

```python
## HTML파일에서 일부분만 선택 str로 저장
html_str ="""
<h1>hello</h1>
<h2>hello <b>world</b></h2>
<a class="A" href="http://www.naver.com">네이버</a>
<a class="A" href="http://www.google.com">구글</a>
<div style='border:2px solid black; background: #2df941;'> GOOD </div>
"""
```

```python
## markup str ==> 객체화 
mbs = BeautifulSoup(html_str, 'html.parser')
```

```python
## [1] 태그객체 전체 가져오기 : BS객체.태그명
print(" h1 태그 전체 : ", mbs.h1)
print("  a 태그 전체 : ", mbs.a)
print("div 태그 전체 : ", mbs.div)
```

```python
## [2] 시작 태그 ~ 끝 태그 사이 텍스트 가져오기
## - .get_string()  X 없는 함수
print("h1 텍스트 :", mbs.h1.string)  # "hello"  (텍스트만 있어서 OK)
print("h2 텍스트 :", mbs.h2.string)  # None     (안에 <b> 태그가 있어서)

##- 일반적인 방법 : Tag객체명.text 또는 Tag객체명.get_text()
print("h1 텍스트 :", mbs.h1.text, mbs.h1.get_text())
print("h2 텍스트 :", mbs.h2.text, mbs.h2.get_text())
print("a  텍스트 :", mbs.a.text,  mbs.a.get_text())
```

```python
## [3] 태그 속성 가져오기 : Tag변수명['속성명'] /  Tag변수명.get('속성명')
## ** 속성 : 태그에 부여된 다양한 옵션
##          시작 태그 내에 속성 설정 
##          [예] <a href="이동URL" target="띄우는 방법"> .... </a>
##               <table style='border: 2px solid red;'> </table>

print("  a 태그의  href 속성값  :", mbs.a['href'])
print("  a 태그의  href 속성값  :", mbs.a.get('href'))

print("div 태그의 style 속성값  :", mbs.div.get('style'))

print("  a 태그의  class 속성값 :", mbs.a['class'])

```
