#### 【 Timestamp 객체 살펴보기 】
- 특정 날짜/시간을 저장하는 클래스
- python의 datetime 모듈과 유사 
- 속성 : 특정 날짜/시간 정보를 추출할 수 있는 속성들 많음
    * 예: .year, .month, .day, .hour ....
- 메서드 : 날짜/시간 변환, 속성값 읽기 등의 것들

```python
## 모듈 로딩 
import pandas as pd 
```

```python
## ==========================================
## Timestamp 인스턴스 생성 및 속성들
## ==========================================
ts1 =pd.Timestamp('2025-11-08 19:25')
ts2 =pd.Timestamp(year=2030, month=9, day=12)

print(f'ts1 => {ts1} {type(ts1)}')
print(f'ts2 => {ts2} {type(ts2)}')

## 속성들 : 인스턴스변수명.속성명
print(f'ts1.year : {ts1.year}')
print(f'ts1.day  : {ts1.day} / {ts1.day_name()}')

print(f'ts2.year : {ts2.year}')
print(f'ts2.day  : {ts2.day} / {ts2.day_name()}')
print(f'ts2.momth: {ts2.month} / {ts2.month_name()}')
```

```python
## 현재 지역의 날짜시간정보 
now_ = ts2.now()
print(f'현재 지역 날짜 시간 : {now_}')

## timestamp => str 
strTS = ts2.ctime()
print(f'timestamp를 str : {strTS}  / {type(strTS)}')
```

==> DatetimeIndex 생성 <hr>

```python
## -----------------------------------------------------
## Index들 ...
## -----------------------------------------------------

## [1] RangeIndex : panda에서 위치 지정한 값으로 구성된 인덱스
sr = pd.Series([11,22,33,44,55])

rangeIdx = sr.index 
print(f'rangeIdx => {rangeIdx}')


## [2] 라벨인덱스 / 인덱스 종류에 따라 다양
sr = pd.Series([11,22,33,44,55], index=['A','B','C','D','E'])

labelIdx = sr.index 
print(f'labelIdx => {labelIdx}')


## [3] 라벨인덱스 / 인덱스 종류에 따라 다양
sr = pd.Series([11,22,33,44,55], index=[7, 8, 9, 10, 11])

labelIdx = sr.index 
print(f'labelIdx => {labelIdx}')
```

```python
## 시간 기준 DatetimeIndex 생성
tsIdx = pd.date_range(start='2025-01-01',
                      freq='2h5min',
                      periods=10)

print(f'tsIdx    => {tsIdx}\n{type(tsIdx)}')
print(f'tsIdx[0] => {tsIdx[0]}\n{type(tsIdx[0])}')
```

```python
## Series 객체 생성 시 날짜시간 인덱스 지정 
tsIdx = pd.date_range(start='2025-01-01',
                      end='2025-01-05',
                      periods=5)
print(tsIdx)
sr = pd.Series([11,22,33,44,55], index = tsIdx)
sr
```
