```python
## 예제 6-3
def calc_stats(col) :
    max_val = col.max()
    min_val = col.min()
    mean_val = col.mean()   
    median_val = col.median()

    return pd.Series([max_val, min_val, mean_val, median_val], index=['Max', 'Min', 'Mean', 'Median'])

result_df = df.apply(calc_stats, axis=0)
result_df
```

```python
result_sr = df.apply(lambda x : x.max()- x.min())
result_sr
```

```python
def calc_diff_avg(row):
    diff = row.max() - row.min()
    avg = row.mean()
    return pd.Series([diff, avg], index=['차이', '평균'])

result_sr2 = df.apply(calc_diff_avg, axis = 1)
result_sr2
```

```python


result_df3 = df.apply(lambda row, multiplier: pd.Series([(row.max() - row.min()) * multiplier, row.mean()],
                                                         index=['차이', '평균']),multiplier = 2, axis=1)
result_df3
```

```python
df1 = titanic.loc[:,['age','fare']]
df2 = titanic.loc[:, ['embark_town','embarked']]

def missing_value(df) :
    return df.isnull()  

result_df = df1.pipe(missing_value)
```

```python
def missing_count(df) :
    return missing_value(df).sum()

result_mc = df.pipe(missing_count)
result_mc
```

```python
def totoal_number_ms(df):
    return missing_count(df).sum()

result_ms = df.pipe(totoal_number_ms)
result_ms
```

```python
# chain method - 함수 활용
def extract_initial(df):
    df['town_initial'] = df['embark_town'].str[0]
    return df
    

def verify_initial(df):
    df['verified'] = df['embarked'] == df['town_initial']
    return df
    
verify_initial(extract_initial(df2))
```

```python
df2.pipe(extract_initial).pipe(verify_initial)
```
