### [데이터 시각화]
- 데이터셋 : iris.csv
- 해결문제 : 품종별 시각화에 적합한 특성/속성/컬럼 선정
- 출력결과 : 선정된 특성을 기반으로 품종 분류한 것 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import numpy as np


dataDF = pd.read_csv('../Data/iris.csv')
dataDF.head(9)

dataDF.set_index('variety')
df_mean = dataDF.groupby('variety')['Age'].mean()
display(df_mean)


x = np.arange(len(df_mean))
W, W2 = 0.8, 0.8*2

plt.bar(x, df_mean.loc["Versicolor"],width=W,label='Versicolor')
plt.bar(x + W, df_mean.loc['Virginica'],width=W, label = 'Virginica')
plt.bar(x+W2, df_mean.loc['Setosa'],width=W, label ='Setosa')

plt.xticks(x+W, df_mean.columns)
plt.xlabel('특징(feature)')
plt.ylabel('평균값')
plt.title('Iris 품종별 특징 평균 비교')
plt.legend()
plt.show()
display(df_mean.columns)


```
